Removing the outliers using z-score

```python
mean = np.mean(df['newspaper'])
std = np.std(df['newspaper'])
print((df['newspaper']-mean)/std)
```

```python
df['newspapers'] = (df['newspaper']-mean)/std
print(df.head(5))
```

```python
# create empty list
outlier_indexes = []

# add index numbers of outliers to list
outlier_indexes.extend(df.index[df['newspapers']>3].tolist())
outlier_indexes.extend(df.index[df['newspapers']<-3].tolist())

print(outlier_indexes)
```
